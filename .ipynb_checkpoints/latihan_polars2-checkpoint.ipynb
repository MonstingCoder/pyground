```python
# REQUIRE_ONCE :

data = r'house price/train.csv'
df = read_csv(data)
df = pl.from_pandas(df)

data = r'hapusAja.csv'
df.write_csv(data)
```

```python
df = pl.read_csv(
  data,
  try_parse_dates=True,
  n_threads=1,
  low_memory=True
)
```

```python
with pl.Config() as plc:
  plc.set_tbl_rows(len(df.columns) + 1)
  display(df.describe().with_columns(
    pl.col('statistic')
    .str.to_uppercase()
  ).transpose(include_header=True).with_columns(
    pl.col('column')
    .str.to_uppercase()
  ))
```
